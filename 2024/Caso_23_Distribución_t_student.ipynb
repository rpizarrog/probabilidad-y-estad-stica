# Fundamento teórico

## Características de T Student

La distribución *t student* tiene estas características:

-   Tiene forma de montículo o campana de gauss y es simétrica alrededor de $t = 0$, igual que $z$ la normal estándar.

- Los valores de $t$ de la distribución *t student* serán los mismos que los valores $z$ de la norma estándar, es decir media igual a cero $\mu= 0$ y desviación estándar alrededor de 1 $\sigma≈1$, estos representados en eje de las x's.

-   Es más variable que $z$, con "colas más pesadas"; esto es, la curva $t$ no aproxima al eje horizontal con la misma rapidez que $z$. Esto es porque el estadístico $t$ abarca dos cantidades aleatorias, $\bar{x}$ y $s$, en tanto que el estadístico $z$ tiene sólo la media muestral, $\bar{x}$. Ver curvas de *t Student* y Normal Estándar $z$.

-   La forma de la distribución $t$ depende del tamaño muestral $n$. A medida que $n$ aumenta, la variabilidad de $t$ disminuye porque la estimación $s$ de $\sigma$ está basada en más y más información.

-   Cuando $n$ sea infinitamente grande, las distribuciones $t$ y $z$ son idénticas. [@mendenhall2010].


El divisor $(n-1)$ en la fórmula para la varianza muestral $s^2 = \sum(\frac{x_i-\bar{x}}{n-1})$ se denomina número de grados de libertad **(df)** asociado con $s^2$ determina la forma de la distribución $t$.

El origen del término grados de libertad es teórico y se refiere al número de desviaciones independientes elevadas al cuadrado en $s^2$ existentes para estimar $\sigma^2$.

Estos grados de libertad pueden cambiar para diferentes aplicaciones y como especifican la distribución *t* correcta a usar, es necesario recordar que hay que calcular los grados de libertad correctos para cada aplicación. [@mendenhall2010].

Si la muestra tiene un valor de $t$ en el rango del nivel de confianza entonces se acepta la hipótesis de lo contrario de rechaza.

## ¿Para qué utilizar t student?

La distribución *t de Student* se utiliza principalmente en inferencia estadística cuando se trabaja con muestras pequeñas (o muestras de las cuales no conocemos la desviación estándar de la población).

- Sirve para etimar intervalos de confianza para la media: Cuando se desea calcualr el intervalo de confianza para la media de una población y la muestra es pequeña, se utiliza la distribución *t student* en lugar de la distribución normal. Esto es especialmente importante cuando la desviación estándar de la población es desconocida.

- Para prueba de hipótesis para la media: Al realizar pruebas de hipótesis sobre la media de una población con una muestra pequeña y desconociendo la desviación estándar de la población, se utiliza la distribución *t student*.

- Para regresión lineal: En el contexto de la regresión lineal, la distribución *t student* se utiliza para realizar pruebas sobre los coeficientes de la regresión y para construir intervalos de confianza para estos coeficientes.

- Comparación de medias de dos Muestras: Cuando se comparan las medias de dos muestras independientes y ambas muestras son pequeñas, se utiliza la distribución *t student* en lugar de la distribución normal.

La distribución $t$ tiene colas más pesadas en comparación con la distribución normal, lo que refleja la mayor incertidumbre asociada con la estimación de la desviación estándar de la población a partir de una muestra pequeña. Conforme el tamaño de la muestra aumenta, la distribución t converge hacia la distribución normal estándar.

## Densidad de una t student

### Densidades con valores de t

$$
f(x) = \frac{\frac{\gamma(gl + 1)}{2}}{\sqrt(gl \cdot \pi)\cdot \gamma(gl/2)} \cdot (1 + \frac{t^2}{gl})^{\frac{-(gl+1)}{2}} \therefore
$$

$$
f(x) = \frac{\frac{gamma(gl + 1)}{2}}{\sqrt(gl \cdot \pi)\cdot gamma(gl/2)} \cdot (1 + \frac{t^2}{gl})^{\frac{-(gl+1)}{2}} \therefore
$$

De acuerdo con la literatura consultada, la función gamma, denotada comúnmente por Γ(x), es una función matemática que generaliza el concepto de factorial para números reales y complejos. Para números enteros positivos, Γ(x) es equivalente al factorial de $!(x−1)$

La función gamma está definida por la integral:

$$
Γ(x) = ∫_{0}^{∞}t^{x-1}e^{-t}dt
$$

En python se utilzia la función *t.pdf* de la bilioteca o librería *scipy.stats import t*.

## Convetir valores x's a valores *t student*

Para convertir valores de una variable aleatoria x a valores t student se utiliza la siguiente fórmula:

$$
t = \frac{x-\mu}{\sigma}
$$

Los valores de t y z son similares en concepto, ya que ambos son formas de estandarizar una distribución para facilitar la comparación con una distribución estándar.

Sin embargo, no son exactamente lo mismo y no siempre son intercambiables, especialmente debido a las diferencias en los contextos en los que se utilizan y en sus fórmulas específicas.



# Desarrollo



## Cargar librerías


```python
import math
import numpy as np

import matplotlib.pyplot as plt
import pandas as pd

from scipy.stats import norm
from scipy.stats import t
```

## Cargar funciones

```python
# Densidad t student
# Recibe un valor numérico de una variabe y los grado de libertdad
def t_student_dens(t, gl):
  numerator = math.gamma((gl + 1) / 2)
  denominator = math.sqrt(gl * math.pi) * math.gamma(gl / 2)
  densidad = numerator / denominator * (1 + t**2 / gl)**(- (gl + 1) / 2)
  return densidad

# Función para visualizar la densidad de probabilidad de una distribución normal
def f_visualizar_dens_normal(valores, titulo):
    # Se calcula la media de los valores
    media = np.mean(valores)

    # Se calcula la desviacióne st´dnar de los valores
    desv_std = np.std(valores)

    # Crear un rango de valores para la crear una densidad
    x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, 1000)

    # Calcular la PDF de la distribución normal
    pdf = 1 / (desv_std * np.sqrt(2 * np.pi)) * np.exp(-(x - media)**2 / (2 * desv_std **2))

    # Visualizar la PDF
    if titulo == 'Normal Z':
      plt.plot(x, pdf, color = 'green',label=titulo + '(media:{}; ds:{})'.format(np.round(media, 2), np.round(desv_std, 2)))
      plt.xlabel('Valor de z')
    else:
      plt.plot(x, pdf, color = 'blue',label=titulo + '(media:{}; ds:{})'.format(np.round(media, 2), np.round(desv_std, 2)))
      plt.xlabel('Valor de x')
    plt.title('Función de Densidad - Distribución ' + titulo)
    plt.ylabel('Densidad ')
    plt.legend(loc='upper right')

    # plt.show()

# Función para visualizar la densidad de probabilidad de una distribución normal
def f_visualizar_dens_t(valores, gl):
    # Se calcula la media de los valores
    media = np.mean(valores)

    # Se calcula la desviacióne st´dnar de los valores
    desv_std = np.std(valores)

    # Crear un rango de valores para la crear una densidad
    x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, 1000)

    # Calcular la PDF de la distribución normal
    pdf = t.pdf(x, gl)

    # Visualizar la PDF
    plt.plot(x, pdf, color = 'yellow', label='t student' + '(me:{}; ds:{}); gl:{}'.format(np.round(media, 2), np.round(desv_std, 2), gl))
    plt.title('Función de Densidad - Distribución ' + 't student')
    plt.xlabel('Valor de t')
    plt.ylabel('Densidad ')
    plt.legend(loc='upper right')

    # plt.show()
def f_visualizar_dens_area_t(valores, gl, a, b, prob=0, v_color = 'red', titulo='t student'):
    # Se calcula la media de los valores
    media = np.mean(valores)

    # Se calcula la desviacióne estándar de los valores
    desv_std = np.std(valores)

    # Crear un rango de valores para la PDF
    x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, 1000)

    # Calcular la PDF de la distribución normal
    pdf = t.pdf(x, gl)

    # La probabilidad
    prob  = t.cdf(b, gl) - t.cdf(a, gl)
    # Visualizar la PDF
    plt.plot(x, pdf, label=titulo+':(me:{}; ds:{}; gl:{})'.format(np.round(media, 2), np.round(desv_std, 2), gl))
    if prob == 0:
      plt.fill_between(x, 0, pdf, where=(x >= a) & (x <= b), alpha=0.3, color=v_color, label='Área bajo la curva')
    else:
      plt.fill_between(x, 0, pdf, where=(x >= a) & (x <= b), alpha=0.3, color=v_color, label='Área ['+str(np.round(a, 2)) + " a "+str(np.round(b, 2))+"]="+str(np.round(prob * 100, 2))+"%")

    plt.title('Función de Probabilidad (PDF) - '+titulo)
    plt.xlabel('Valor de t')
    plt.ylabel('Densidad ')
    plt.legend(loc='upper right')
    # plt.grid(True)
    #plt.show()

def f_visualizar_VE_desv_std_t(valores, gl):
  # Se calcula la media de los valores
  VE = media = np.mean(valores)

  # Se calcula la desviacióne st´dnar de los valores
  desv_std = np.std(valores)

  # Crear un rango de valores para la PDF
  x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, 1000)

  # Calcular la PDF de la distribución normal
  pdf = 1 / (desv_std * np.sqrt(2 * np.pi)) * np.exp(-(x - media)**2 / (2 * desv_std ** 2))
  plt.plot(x, pdf, label='Normal:(me:{}; ds:{})'.format(np.round(media, 2), np.round(desv_std, 2)))

  # Agregar una línea punteada
  plt.axvline(x=VE, color='red', linestyle='--', label='VE='+str(np.round(VE, 2)))
  plt.axvline(x=VE - desv_std, color='blue', linestyle='--', label='Desv. Std='+str(np.round(desv_std, 2)))
  plt.axvline(x=VE + desv_std, color='blue', linestyle='--')

  # La probadilidad de VE ± desv_std
  a = VE - desv_std
  b = VE + desv_std
  prob  = norm.cdf(b, media, desv_std) - norm.cdf(a, media, desv_std)

  plt.fill_between(x, 0, pdf, where=(x >= VE - desv_std) & (x <= VE + desv_std), alpha=0.3, color='red', label='Área '+str(np.round(VE - desv_std, 2)) + "-"+str(np.round(VE + desv_std, 2))+ "="+str(np.round(prob, 2)))


  plt.title('VE y desviación estándar de los datos. Distribución Normal')
  plt.xlabel('Valor de x')
  plt.ylabel('Densidad de Probabilidad')
  plt.legend(loc='upper right')
  # plt.grid(True)
  plt.show()

# Se reciben valores de x y se devuelven valores en z
def f_transformar_x_z(x, media, desv_std):
  x = (x - media) / desv_std
  return x

# Se reciben valores de x y se devuelven valores en t
def f_transformar_x_t(x, media, desv_std):
  x = (x - media) / desv_std
  return x
```

## Visualización de densidad de t sudent comparada con normal y noraml z

Se crean $n=30$ números aleatorios de una distribución  normal con media igual a $\mu = 80$ y desviación estándar de $\sigma = 5$ (las calificaciones de casos anteriores), se simula la densidad de la distribución  normal para esos números, luego se traansforman a valores $z$ y a valores $t$ respectivamente.

Al tener los tres conjuntos de datos de las tres densidades (de la normal, de la normal $z$ y de $t student$) se visualizan las tres densidades y se compara visualmente el comportamiento de las respectivas curvas.



```python
media = 80
desv_std = 5
n = 30
gl = n-1
xs = norm.rvs(loc = media, scale = desv_std, size = n)
print(np.round(xs, 4))

media = np.mean(xs) # media actuliada
desv_std = np.std(xs) # desviación estándar actualizada

print()
print("Los estadísticos a partir de los datos den ser similares a 80 y 5 respectivamente")
print ("La media de los datos = ", np.round(media, 2))
print ("La desviación estándar de los datos = ", np.round(desv_std, 2))
```

### Transformar a valores z

```python
# Transformar a z
zs = f_transformar_x_z(xs, media, desv_std)
print ("Valores de z")
print (zs)
```

## Transformar a valores ts

Los valores de $t's$ serán los mismos valores que los valores de $z's$.



```python
# Transformar a t
ts = f_transformar_x_t(xs, media, desv_std)
print ("Valores de t")
print (ts)
```

## Probabilidad en *t studen*

### $F(t \le -1)$ a $gl=10$

¿Cuál es la probabilidad en una distribución *t student* de $t \le-1$ a $10$ grados de libertad $gl=10$?; $F(t \le -1)$

Se calcula y se visualiza la probabilidad y llamando la función *f()*.

Los límites son infinio negativo $-∞$ a la izquierda e infinito positivo $∞$ a la derecha por tal razón las colas nunca tocan el eje de las x en el plano.

```python
probabilidad = t.cdf(x = -1, df = 10) -  t.cdf(x = -np.inf, df = 10)
print ("La probabilidad es de : ", probabilidad,  " o ", np.round(probabilidad *100, 4), "%")
f_visualizar_dens_area_t(valores = ts, gl = gl, a = -np.inf, b = -1)

```
