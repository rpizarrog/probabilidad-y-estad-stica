# Marco teórico

## Operación Union U.

El conjunto de todos los elementos que pertenecen a A o a B, o tanto a A como a B, se llama la unión de A y B y se escribe A ∪ B.

En la unión si hay elementos repetidos sólo se deja uno de ellos.

![](images/union%20conjuntos.png)

## Operación intersección ∩

El conjunto de todos los elementos que pertenecen simultáneamente a A y B se llama la intersección de A y B y se escribe A ∩ B.

![](images/interseccion-01.png)

## Operación diferencia -

El conjunto que consiste en todos los elementos de A que no pertenecen a B se llama la diferencia de A y B y se escribe A -- B.

![](images/diferencia.png)

## Operación complemento [C ó ´]

Son todos los conjuntos con los elementos que no están en A y se escribe A' ó C A. Son todos los elementos que faltan y que no están en A para complementar todo el espacio muestral.

![](images/complemento-01.png)

# Desarrollo

## Cargar librerías

```python
# Cargar librerías
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
from matplotlib_venn import venn3
```

## Cargar datos


```python
url = "https://raw.githubusercontent.com/rpizarrog/probabilidad-y-estad-stica/master/2023/datos/alumnos_actividades_extraescolares.csv"
datos = pd.read_csv(url)



print("Los nombres de datos (alumnos)")

print(datos.head(70))
```

## Crear conjuntos de varias disciplinas
* FUT : FUTBOL
* BAS : BASQUETBOL
* VOL : VOLEYBOL
* TEN : TENIS
* AJE : AJEDREZ
* ATL : ATLETISMO
* DAN : DANZA
* RON : RONDALLA
* TEA : TEATRO
* NAT : NATACIÓN





```python
random.seed(2024)

nombres = datos["nombres"].tolist()
N = len(nombres)

# Barajar los nombres aleatoriamente
random.shuffle(nombres)


# Barajar los nombres aleatoriamente
random.shuffle(nombres)

# Hacerlo conjunto la lista de nombres
nombres = set(nombres)
print("Todos los nombres")
print(nombres)
print("\n")

# FUT FUTBOL
FUT = set(random.sample(nombres,  20))
print("Conjunto FUT: FUTBOL")
print(FUT)
n = len(FUT)
print("La probabilidad de elegir a un alumno que sea de FUTBOL es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")
print("\n")

# BAS BASQUETBOL
BAS = set(random.sample(nombres,  12))
print("Conjunto BAS: BASQUETBOL")
print(BAS)
n = len(BAS)
print("La probabilidad de elegir a un alumno que sea de BASQUETBOL es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")
print("\n")


# VOL VOLEYBOL
VOL = set(random.sample(nombres,  12))
print("Conjunto VOL: VOLEYBOL")
print(VOL)
n = len(VOL)
print("La probabilidad de elegir a un alumno que sea de VOLEYBOL es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")
print("\n")

# TEN TENIS
TEN = set(random.sample(nombres,  10))
print("Conjunto TEN: TENIS")
print(TEN)
n = len(TEN)
print("La probabilidad de elegir a un alumno que sea de TENIS es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")
print("\n")

# AJE AJEDREZ
AJE = set(random.sample(nombres,  8))
print("Conjunto AJE: AJEDREZ")
print(AJE)
n = len(AJE)
print("La probabilidad de elegir a un alumno que sea de AJEDREZ es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")
print("\n")

# ATLETISMO
ATL = set(random.sample(nombres,  16))
print("Conjunto ATL: ATLETISMO")
print(ATL)
n = len(ATL)
print("La probabilidad de elegir a un alumno que sea de ATLETISMO es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")
print("\n")

# DANZA
DAN = set(random.sample(nombres,  12))
print("Conjunto DAN: DANZA")
print(DAN)
n = len(DAN)
print("La probabilidad de elegir a un alumno que sea de DANZA es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")
print("\n")

# RONDALLA
RON = set(random.sample(nombres,  14))
print("Conjunto RON: RONDALLA")
print(RON)
n = len(RON)
print("La probabilidad de elegir a un alumno que sea de RONDALLA es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")
print("\n")

# TEATRO
TEA = set(random.sample(nombres,  9))
print("Conjunto TEA: TEATRO")
print(TEA)
n = len(TEA)
print("La probabilidad de elegir a un alumno que sea de TEATRO es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")
print("\n")

# NAT NATACION
NAT = set(random.sample(nombres,  10))
print("Conjunto NAT: NATACION")
print(NAT)
n = len(NAT)
print("La probabilidad de elegir a un alumno que sea de NATACION: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")
print("\n")

```

## Unión de conjuntos U

### Unión de FUTBOL con ATLETISMO




```python
# FUTBOL U ATLETISMO
resultado = FUT.union(ATL)
print(resultado)
n = len(resultado) # resulado de la operación de conjuntos
print("La probabilidad de elegir a un alumno que sea de FUTBOL O ATLETISMO es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")


# Diagrama de Venn
venn2([FUT, ATL], ('FUTBOL', 'ATLETISMO'))

# Mostrar los elementos de cada conjunto en el título
plt.title("Conjuntos FUTBOL y ATLETISMO\nFUT: {}\nATL: {}".format(FUT, ATL))
plt.show()
```

### Unión de TEATRO con RONDALLA


```python
# TEATRO U RONDALLA
resultado = TEA.union(RON)
print(resultado)
n = len(resultado) # resulado de la operación de conjuntos
print("La probabilidad de elegir a un alumno que sea de TATRO O RONDALLA es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")


# Diagrama de Venn
venn2([TEA, RON], ('TEATRO', 'RONDALLA'))

# Mostrar los elementos de cada conjunto en el título
plt.title("Conjuntos TEATRO y RONDALLA\nTEA: {}\nRON: {}".format(TEA, RON))
plt.show()
```

## AJEDREZ U TENIS


```python
# AJEDREZ U TENIS
resultado = AJE.union(TEN)
print(resultado)
n = len(resultado) # resulado de la operación de conjuntos
print("La probabilidad de elegir a un alumno que sea de AJEDREZ O TENIS es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")

# Diagrama de Venn
venn2([AJE, TEN], ('AJEDREZ', 'TENIS'))

# Mostrar los elementos de cada conjunto en el título
plt.title("Conjuntos AJEDREZ y TENIS\nAJE: {}\nTEA: {}".format(AJE, TEA))
plt.show()
```

### Intersección de conjuntos ⋂
#### Intersección  de FUTBOL con ATLETISMO

```python
# FUTBOL ⋂ ATLETISMO
resultado = FUT.intersection(ATL)
print(resultado)
n = len(resultado) # resulado de la operación de conjuntos
print("La probabilidad de elegir a un alumno que sea de FUTBOL y ATLETISMO es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")

# Diagrama de Venn
venn2([FUT, ATL], ('FUTBOL', 'ATLETISMO'))

# Mostrar los elementos de cada conjunto en el título
plt.title("Conjuntos FUTBOL y ATLETISMO\nFUT: {}\nATL: {}".format(FUT, ATL))
plt.show()
```

### Intersección de BASQUETBOL con VOLEYBOL

```python
# BASQUETBOL ⋂ VOLEYBOL
resultado = BAS.intersection(VOL)
print(resultado)
n = len(resultado) # resulado de la operación de conjuntos
print("La probabilidad de elegir a un alumno que sea de BASQUETBOL y VOLEYBOL es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")

# Diagrama de Venn
venn2([BAS, VOL], ('BASQUETBOL', 'VOLEYBOL'))

# Mostrar los elementos de cada conjunto en el título
plt.title("Conjuntos BASQUETBOL y VOLEYBOL\nBAS: {}\nVOL: {}".format(BAS, VOL))
plt.show()
```

# Intersección AJEDREZ, RONDALLA Y TEATRO

¿cuál es la probabildia de elegir a un alumno que esté inscritp en AJEDREZ RONDALLA Y TEATRO?


```python
# AJEDREZ ⋂ RONDALLA ⋂ TEATRO
resultado = (AJE.intersection(RON)).intersection(TEA)
print(resultado)
n = len(resultado) # resulado de la operación de conjuntos
print("La probabilidad de elegir a un alumno que sea de AJEDREZ, RONDALLA Y TEATRO es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")

print(AJE)
print(RON)
print(TEA)
# Diagrama de Venn
venn3([AJE, RON, TEA], ('AJEDREZ', 'RONDALLA', 'TEATRO'))

# Mostrar los elementos de cada conjunto en el título
plt.title("Conjuntos AJEDREZ, Y RONDALLA Y TEATRO\nAJE: {}\nRON: {}\nTEA: {}".format(AJE, RON, TEA))
plt.show()
```

### Complemento de AJEDREZ

Se representa C(AJEDREZ) ó AJEDREZ '

¿Cuál es la probabilidad de que un alumno no juegue AJEDREZ, o no esté inscrit en AJEDREZ?

Para encontrar el complemento se puede restar de universo de nombres de alumnos los que están en AJEDREZ.






```python
# alumnos - AJEDREZ = Complemento de AJEDREZ
resultado = (nombres - AJE)
print(resultado)
n = len(resultado) # resultado de la operación de conjuntos
print("La probabilidad de elegir a un alumno que no sea de AJEDREZ: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")

# Diagrama de Venn
venn2([nombres, AJE], ('nombres', 'AJEDREZ'))

# Mostrar los elementos de cada conjunto en el título
plt.title("Conjuntos nombres y AJEDREZ \nnombres: {}\nAJE: {}".format(nombres, AJE))
plt.show()
```
