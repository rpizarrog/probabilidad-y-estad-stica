# Fundamento teórico

La prueba de hipótesis es un método estadístico que se utiliza para tomar decisiones sobre las características de una población, basándose en la información obtenida de una muestra de esa población.

¿Existe relación de prueba de hipótesis con intervalos de confianza?

La relación que existe entre los intervalos de confianza y la prueba de hipótesis es tal, que precisamente los valores de prueba deben estar en la zona de intervalo de confianza determinado y permite con ello hacer contraste de valores de prueba para aceptar o rechazar alguna hipótesis.

La inferencia estadística basada en lograr la resolver la naturaleza de afirmaciones contradictorias está relacionada con la prueba de hipótesis.

Estas afirmaciones contradictorias, tiene que ver con el concepto de declaración de hipótesis. A lo que da lugar el concepto de hipótesis estadística que lo define con claridad.

Una hipótesis estadística o simplemente una hipótesis es una afirmación o aseveración sobre el valor de un solo parámetro (característica de una población o característica de una distribución de probabilidad)” (Devore, 2018, pág. 253).

Una hipótesis estadística es una aseveración o conjetura con respecto a una o más poblaciones” (Mendenhall, Beaver, & Beaver, 2010, pág. 321).


Un contraste o prueba de hipótesis, consiste en plantear dos hipótesis, la hipótesis nula $(H_0)$ y la hipótesis alternativa $(H_1)$, y luego recopilar evidencia en forma de datos para determinar si hay suficiente evidencia para rechazar la hipótesis nula en favor de la hipótesis alternativa.






## Fórmula para calcular el valor de z para contraste o prueba contra z críticos

Para calcular $z$ a contrastar se utiliza la fórmula siguiente:

$$
z = \frac{{\bar{x} - \mu}}{{\sigma / \sqrt{n}}}
$$

* $z$ es el valor de prueba de z a contrastar contra valores críticos.
* $\bar{x}$ es la media de la muestra.
* $\mu$ es  la media estimada de la población.
* $\sigma$ es la desviación estándar de la población conocida.
* $n$ es el tamaño de la muestra.

## Proceso para prueba de hipótesis

Proceso para realizar prueba de hipótesis:

1. Formulación de hipótesis: Se establecen la hipótesis nula $(H_0)$, que generalmente afirma que no hay efecto o diferencia, y la hipótesis alternativa $(H_1)$, que afirma lo contrario.

2. Selección de nivel de significancia: Se elige un nivel de significancia (α) que representa la probabilidad de cometer un error de tipo I al rechazar la hipótesis nula cuando es verdadera. Un valor común para alpha es $α$ es $0.05$.

  En casos anteriores, al identificar intervalos de confianza se requería el nivel de confianza para determinar el valor de alfa.

3. Selección de prueba estadística: Se selecciona una prueba estadística apropiada basada en el tipo de datos y la naturaleza de la hipótesis.

4. Cálculo del estadístico de prueba: Se calcula el estadístico de prueba a partir de los datos de la muestra.

5. Decisión: Se compara el valor del estadístico de prueba con un valor crítico basado en la distribución de probabilidad bajo la hipótesis nula, o se calcula un valor $p$ que representa la probabilidad de obtener un estadístico de prueba al menos tan extremo como el observado si la hipótesis nula es verdadera.

6 Conclusión: Si el valor p es menor que α, se rechaza la hipótesis nula en favor de la hipótesis alternativa. De lo contrario, no se puede rechazar la hipótesis nula.

## Escenarios de declaración de hipótesis

Ejemplos de declaración de hipótesis. Par declarar hipótesis se utiliza la expresión $H_0$ para designar hipótesis nula  la expresión $H_1$ o $H_a$ para idenificar la hipótesis alternativa.

La hipótesis nula significa una negación a una afirmación que pretende cambiar lo ya conocido. La afirmación es producto de una pregunta de investigación, lo que da lugar a la hipótesis alternativa y la hipótesis nula es la negación.

La hipótesis nula  hay que demostrar estadísticament que se pueda rchazar o seguir aceptando.

### Ejemplo 1:

En el escenario del acusado, la pregunta de investigción sería: ¿es el acusado culpable?, la  $H_0$  es que el acusado es inocente, "*El acusado no es culplable*"; por consiguiente, la  $H_1$  no es inocente y por consecuencia, "*El acusado es culpable*".

$$
H_0:  \text{Acusado no es  culplable}
\\acusado=inocente
$$

$$
H_1:  \text{Acusado es culplable}
\\acusado ≠inocente
$$

Estadísticamente se tendría que tener elementos cuantificables de la variable inocente que se contrastarían con los elementos de acusaciones para aceptar o rechazar la hipótesis nula.

### Ejemplo 2:

* En un primer escenario de este ejemplo, es estudiar la edad promedio de las personas $(μ)$, una hipótesis es suponer que la media de la edad de una población de personas podría ser que en promedio tengan 32 años $(μ=32)$, y la otra decir lo contrario, que la media no es 32 años $(μ≠32)$.

Pregunta de investigación ¿es diferente la edad promedio de las personas de 32 años?, la hipótesis alternativa $H_a$ es la afirmación a esa pregunta: $H_a$: "**la edad promedio de las personas es difrente de 32 años*" la $H_0$ implica la negación; $H_0$: "*la edad promedio de las personas no es de 32 años*".

$$
H_0: \text{La edad promedio no es de 32 años}\\ edad \ne 32
$$

$$
H_1: \text{La edad promedio es de 32 años}\\ edad = 32
$$

* Un segundo escenario pudiera ser que se afirme que la edad promedio (media de la edad en años) pudiera estar por encima o igual a 32 años $(μ≥32)$ y la segunda afirmación sería lo contrario, edad por debajo de 32 años. $(μ<32)$.

$$
H_0: \text{La edad promedio no es mayor o igual que 32 años}\\ edad < 32
$$

$$
H_1: \text{La edad promedio es mayor o igual a 32 años}\\ edad \ge 32
$$

* Un tercer escenario de la edad promedio de una población sería a la inversa del segundo escenario, afirmar que la edad promedio esté por debajo o igual a 32 años $(μ ≤ 32)$ y la segunda afirmación sería lo contrario, edad por encima de 32 años. $(μ > 32)$.

$$
H_0: \text{La edad promedio no es menor o igual que 32 años}\\ edad > 32
$$

$$
H_1: \text{La edad promedio es menor o igual a 32 años}\\ edad \le 32
$$

## Tipos de hipótesis

En estadística se puede declarar tres tipos de hipótesis:

* hipótesis a dos colas, cuando el valor de significancia de *alpha* $α$ se reparte a ambos lados de la curva y conforme a la probabilidad;

* cola a la izquierda cuando toda la probabilidad está del lado izquierdo y el valor del alpha $α$ es igual al complemento del valor de la confiabilidad.

* cola a la derecha cuando toda la probabilidad está del lado derecho y el valor del alpha $α$ es igual al complemento del valor de la confiabilidad.

De acuerdo la prueba de hipótesis se debe comparar un parámetro estimado de la población contra un valor crítico calculado.

La decisión de aceptar la $H_0$ dependerá de la condición de comparación.

|                 | Dos colas        | Cola a la izquierda | Cola a la derecha |
|-----------------|------------------|----------------------|-------------------|
| Acepta          | $H_0$: parámetro=valor | $H_0$: parámetro≥ valor | $H_0$: parámetro≤ valor |
| Rechaza         | $H_a$: parámetro≠valor | $H_0$: parámetro< valor | $H_0$: parámetro> valor |

Convirtiendo esta tabla a una declaración de una hipótesis en donde se busca indagar una media poblacional por ejemplo $μ$ comparada contra un valor de 68 a dos colas, cola izquierda o cola a la derecha.

El operador de igualdad que define si es a dos colas, cola izquierda o cola derecha es el signo de igualdad:
* '=': es a dos colas,
* '≥':   es cola a la izquierda o cola inferior,
* '≤':  es cola a la derecha o cola superior.

|                 | Dos colas        | Cola a la izquierda | Cola a la derecha |
|-----------------|------------------|----------------------|-------------------|
| Acepta          | $H_0$: $\mu$=68 | $H_0$: $\mu$ ≥ 68 | $H_0$: $\mu$ ≤ 68 |
| Rechaza         | $H_a$: $\mu$≠68 | $H_a$: $\mu$< 68 | $H_a$: $\mu$ > valor |


### Prueba de hipótesis contrastando z contra z críticos

En la hipótesis de dos colas significa que se va a aceptar la hipótesis nula $H_0$ siempre y cuando el valor de z calculado se encuentre entre dos valores críticos que son determinados a partir del valor del alfa divido entre dos. $∝⁄2$.

En la hipótesis cola a la izquierda significa que se va a aceptar la hipótesis nula $H_0$ siempre y cuando el valor de $z$ calculado se encuentre por encima del valor crítico a la izuierda que es determinado a partir del valor del alfa $∝$.

En la hipótesis cola a la derecha significa que se va a aceptar la hipótesis nula $H_0$ siempre y cuando el valor de $z$ calculado se encuentre por debajo del valor crítico a la derecha que es determinado a partir del valor del alfa $∝$.

### Prueba de hipótesis contrastando el valor de *p prueba*

En una prueba de hipótesis, el valor $p$ (*p-value* en inglés) es la probabilidad de obtener un resultado igual al observado o más extremo, bajo la suposición de que la hipótesis nula es verdadera. En otras palabras, el valor $p$ es una medida de la evidencia en contra de la hipótesis nula.

Un valor $p$ pequeño sugiere que la evidencia en contra de la hipótesis nula es fuerte, lo que lleva a rechazar la hipótesis nula en favor de la hipótesis alternativa. Por otro lado, un valor $p$ grande sugiere que la evidencia en contra de la hipótesis nula es débil, lo que lleva a no rechazar la hipótesis nula.

El valor estadístico de *p.valor* se adopta para consolidar la decisión de aceptar o rechazar la hipótesis nula $H_0$. El *p.valor* o *p.value* se calcula en términos de una probabilidad como respaldo a la decisión tomada.

El significado del estadístico *p.valor* es la probabilidad de aceptación o rechazo de la hipótesis nula $H_0$.

Por ejemplo si *p.valor* es menor que $0.05$ u otro valor de significancia hay que rechazar la hipótesis nula $H_0$, de lo contrario aceptar la $H_0$.

Toda vez que se determine el valor estadístico de *p.prueba*, este se debe comparar contra un nivel de significancia establecido que comúmente  puede ser 0.05 pero puede ser diferente.

Entonces el algoritmo para aceptar o rechazar $H_0$ a dos colas con *p.value* sería de la siguiente manera:


A un nivel de confianza del 95%
$$
\text{p_prueba = 0.40 y p.value = 0.05$
$$

$$
\text{if (p_prueba < p.value)  Se rechaza H0}
\\ \text{de lo contrario} \\
\text{if (p_prueba ≥ p.value) Se acepta H0}
$$

$$
\text{if (0.40 < 0.05)  Se rechaza H0}
\\ \text{de lo contrario} \\
\text{if (0.40 < 0.05) Se acepta H0}
$$

**Nota importante**:

En la hipótesis a dos colas el valor de $p$ de prueba (*p.prueba*) que es la probabilidad calculalada con respecto a $z$,  se multiplica dos veces esa probabilidad (por ser a dos colas) para se contrastado contra el *p.value*.

En la hipótesis de cola izquierda o derecha el valor de *p.prueba* es el resultado de la probabilidad con respecto a $z$ directamente contrastado contra *p.value*.

## Valores críticos de $z$ para contraste

Los valores críticos de $z$ es el valor de $z$ en el eje horizontal en donde se cruza con la curva de densidad de la distribución $z$ con respeco a una probabilidad. Ver visualizaciones las líneas punteadas en rojo y el cruce con el eje horizontal.

La siguente tabla identifica los valores críticos de $z$ que servirán para contraste.

Estos valores se pueden calcular con la función de python: :

| Confianza | Alfa | Z (Dos colas) $α/2$ | Z (Izquierda)$α$ | Z (Derecha)$1-α$ |

| 0.95      | 0.05 | 1.959964      | -1.644854      | 1.644854     |
| 0.99      | 0.01 | 2.575829      | -2.326348      | 2.326348     |

Para extrar los valores críticos de $Z$ para diferentes niveles de confianza en Python, se puede utilizar la función *ppf()* de la distribución normal estándar de scipy.stats.

Ejemplo Python: a distintos niveles de confianza del 90%, 95% y 99%:



```python
confianza = [0.90, 0.95, 0.99]
alfa = [1 - c for c in confianza]

for c, a in zip(confianza, alfa):
    z_dos_colas = norm.ppf(1 - a / 2)
    z_izquierda = norm.ppf(a)
    z_derecha = norm.ppf(1 - a)
    print(f"Confianza: {c:.2f}, Alfa: {a:.2f}, Z (Dos colas): {z_dos_colas:.4f}, Z (Izquierda): {z_izquierda:.4f}, Z (Derecha): {z_derecha:.4f}")
```

### Ejercicio 1. Peso de estudiantes

El objetivo es conocer si la media poblacional $\mu$ hipotéticamente estimada del peso promedio de estudiantes hombres en cierta escuela de educación superior es $68$ kilogramos.

Una preguna de investigación tentativa asociada al objetivo sería algo como ¿la media del peso de los estudiantes es diferente a 68 kgs?.

Como la pregunta de investigación es una afirmación de la hipótesis alterntativa esta se declara como  $H_a$: *"la media del peso de la población de estudiantes es diferente de 68 kgs."*, la hipótesis nula es la negación de esta hipótesis alternativa $H_0$: *"la media del peso de la población de estudiantes no es diferente de 68 kgs."*.

$$
H_0: \text{la media del peso de la población de estudiantes NO es diferente de 68 kgs.}
$$

$$
H_a: \text{la media de la población de estudiantes SI es diferente de 68 kgs.}
$$

O bien con simbología estadística y/o matemática:

$$
H_0: \mu = 68
$$

$$
H_a: \mu \ne 68
$$

De acuerdo a la declaración de hipótes es una hipótesis ... a dos colas entonces el valor de significancia de alpha es $\alpha/2$ a ambos lados de la curva.

Ahora bien, como de los datos iniciales si se conoce la desviación estándar de una población entonces se utilizará la distribución z para conocer los valores críticos $z$ a la izquierda y z a la derecha que servirán para el contraste o la comparación.

El nivel de confianza, para este ejercicio se establece que sea el $90$% lo cual indica que bajo un nivel de $90$% de confianza la madia del peso de los estudiantes es de $68$ kgs.

Para este ejercicio, de la muestra se necesita saber los datos iniciales: el tamaño de la muestra $n=36$, la media tomada de la muestra del peso de los estudiantes $\bar{x}=67.5$ y la desviación estándar conocida de la población $σ=3.6$.

#### Prueba de hipótesis con z calculado contrastado con z criticos

¿Cuál es el valor de z calculado?

$$
\begin{aligned}
z &= \frac{\bar{x} - \mu}{\sigma/\sqrt{n}} \\
&= \frac{67.5 - 68}{3.6/\sqrt{36}} \\
&= \frac{-0.5}{3.6/6} \\
&= \frac{-0.5}{0.6} \\
&= -0.8333
\end{aligned}
$$

El valor de $z$ calculado de $-0.8333$ está en zona de aceptación mayor a $-z$ critico por la izquierda y menor a $z$ por la derecha, por lo que se anticipa aceptar la $H_0$ de que el peso de los estudiantes es igual a $68$.



```python
# Los datos
H0 = "la media del peso de la población de estudiantes NO es diferente de 68 kgs."
confianza  = 90/100
media_estimada = 68 # Es la media de comparación
media_muestra = 67.5 # Es la media tomada de la muestra

desv_std_poblacion = 3.6 # Es la desviación estándar de lal población conocida
n = 36 # El tamaño de la muestra

decision = f_aceptar_rechazar_ho(confianza, media_estimada, media_muestra, desv_std_poblacion, n, "D")

print (decision, "H0:", H0, "con nivel de confianza del", confianza*100, "%")
```

A cualquier nivel de confianza ya sea $90$%, $95$% o inclusive $99$% se acepta la hipotesis nula a dos colas: "*la media del peso de la población de estudiantes NO es diferente de 68 kgs.*", dado qu eel valor de $z$ calculado está en zona de confianza o de aceptación de $H_0$.

Se puede deducir e inferir que un nivel de confianza del $90$%, la media del peso en kgs. de de la población de estudiante debe estar aproximado en 68 kgs, aquí es donde tiene relación con el intervalo de confianza.
