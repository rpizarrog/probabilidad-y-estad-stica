# Fundamento teórico

El intervalo de confianza para la media poblacional es un rango estimado dentro del cual se espera que se encuentre la verdadera media de una población con un cierto nivel de confianza. Es una medida estadística que proporciona un margen de error alrededor de la estimación de la media muestral.

Cuando se realiza una estimación puntual de la media de una población basada en una muestra, es importante comprender la incertidumbre asociada con esa estimación.

La inferencia estadística, específicamente la toma y predicción de decisiones desempeña un papel muy importante en la vida de casi todas las personas.

Hay muchas formas de tomar estas decisiones o predicciones, algunas son subjetivas y otras son objetivas por naturaleza. La pregunta es ¿qué tan buena son las decisiones y en qué argumentos se basan estas decisiones?

Aun cuando se pueda pensar que su propia capacidad de tomar decisiones es muy buena, la experiencia sugiere que éste puede no ser el caso.

Es la función del estadístico matemático dar métodos de toma de inferencia estadística son mejores y más confiables que únicamente cálculos subjetivos.[@mendenhall2010].

La inferencia estadística se encarga de apoyar el proceso de toma decisiones o predicciones acerca de parámetros. Tal vez estos parámetros de interés sean la media poblacional $\mu$, la desviación estándar poblacional $\sigma$ o la proporción de una población $p$.

Para este caso en los ejercicios mostrados mas adelante, se describen estimación de las medias poblacionales.

Cabe recordar que si se trata de mediciones poblacionales estos se llaman parámetros y si se trata de mediciones muestrales estos se llaman estadísticos.

- $población = parámetros$
- $muestra = estadísticos$

De acuerdo a [@mendenhall2010] estimación implica predecir el valor del parámetro.

Para estimar el valor de un parámetro poblacional, se puede usar datos de la muestra en la forma de un estimador. Los estimadores se determinan usando información de las observaciones muestrales y, en consecuencia, por definición son también estadísticas.

Por definición, un estimador es una regla, generalmente expresada como fórmula, que establece cómo calcular una estimación basada en información de la muestra.

Se pueden encontrar dos tipos de estimaciones

- Estimaciones puntuales que significa que con base en datos muestrales, se calcula un solo número o un estimador puntual.

- Estimaciones por intervalos que implica encontrar dos valores numéricos que se calculan para formar un intervalo dentro del cual se espera esté el parámetro poblacional. Los intervalos son el interés de este caso.

Ahora bien, es posible intuir que la distribución muestral de un estimador está centrada alrededor del parámetro que se trate de estimar, pero todo lo que se tiene es la estimación calculada de las $n$ mediciones contenidas en la muestra.

¿A qué distancia del verdadero valor del parámetro estará esta estimación? ¿Qué tan diferente es el valor del estadísticos con respecto al parámetro?

La distancia entre la estimación y el verdadero valor del parámetro se denomina error de estimación. [@mendenhall2010].

Se puede suponer que los tamaños muestrales son siempre grandes y que los estimadores provienen de distribuciones muestrales que pueden ser aproximadas por una distribución normal por el teorema del límite central.

La variabilidad del estimador se mide usando este error estándar y está determinado por:

$$
ES = \frac{\sigma}{\sqrt{n}}
$$

Este error estándar se incorpora y se utiliza para estimar intervalos de confianza como la media poblacional.

Ahora bien, el objetivo de la estimación por intervalo es aportar información de qué tan cerca se encuentra la estimación puntual, obtenida de la muestra, del valor del parámetro poblacional [@anderson2008].

Se identifican las fórmulas para estimar intervalos de confianza con desviación estándar de la población conocida y no conocida respectivamente.

El intervalo de confianza ofrece esa información. Por ejemplo, un intervalo de confianza del 95% indica que si se repite el proceso de muestreo y se construye intervalos de confianza para la media poblacional muchas veces, se espera que aproximadamente el 95% de esos intervalos contengan la verdadera media poblacional.

El cálculo del intervalo de confianza para la media poblacional a menudo se realiza utilizando la distribución t de Student cuando se trabaja con muestras pequeñas o cuando la desviación estándar de la población no es conocida.

En el caso de muestras grandes y cuando la desviación estándar de la población es conocida, la distribución normal estándar (z) a menudo se utiliza para calcular el intervalo de confianza.

¿entonces cuándo usar $z$ y cuándo usar $t$?

## Fórmulas para intervalos de confianza de la media poblacional

### Si SI se conoce la desviación estándar de la población usar:
$$
IC = \bar{x} \pm z \cdot ES \therefore
$$

$$
IC = \bar{x} \pm z \cdot \frac{\sigma}{\sqrt{n}}
$$
- $\bar{x}$ es la media muestral.
- z es el valor de z calculado.
- ES = error estándar $\frac{\sigma}{\sqrt{n}}$
- $\sigma$ es la desviación poblacional conocida.
- $n$ tamaño de la población.

Para calcular valores z:
$$
z = \frac{x_i-\bar{x}}{\sigma}
$$
- $x_i$ es un valor de variable aleatoria x.
- z es el valor calculado o transforamdo de $x_i$.

Para calcular los valores *críticos de z* dependiendo del nivel de confianza se utiliza la función de probabilidad inversa a la probabilidad acumulada *norm.ppf()* de la librería *scipi.stats*.

Esta función *norm.ppf()* devuelve valores de $z$ dependiendo del nivel de confianza. Los valores de $z$ serán a dos colas conforme al valor de confianza y al valor de alfa $\alpha$.

$$
\alpha = (1 - confianza) / 2
$$


### Si NO se conoce la desviación estándar de la población usar:

$$
IC = \bar{x} \pm t \cdot ES \therefore
$$

$$
IC = \bar{x} \pm t \cdot \frac{S}{\sqrt{n}}
$$

- $\bar{x}$ es la media de la muestra
- $ES$ es el error estándar calculado por $\frac{S}{\sqrt{n}}$.
- $t$ es el valor de t calculado o transformado
- $S$ es la desviación estándar de la muestra
- $n$ es el tamaño de la muestra

Para calcular los *valores críticos de t* dependiendo del nivel de confianza se utiliza la función de probabilidad inversa a la probabilidad acumulada *t.ppf()* de la librería *scipi.stats*.

Esta función *t.ppf()* devuelve valores de $t$ dependiendo del nivel de confianza. Los valores de $t$ serán a dos colas conforme al valor de confianza y al valor de alfa $\alpha$.

$$
\alpha = (1 - confianza) / 2
$$


# Desarrollo



## Cargar librerías



```python
import numpy as np
from scipy.stats import norm
from scipy.stats import t

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Cargar funciones



```python
def f_intervalo_confianza_z(media_m, desv_std_p, n, confianza):
  # Calcular el error estándar de la media (SEM)
  es = desv_std_p / np.sqrt(n)  # error estándar desv_std/sqrt(n)
  # Calcular el valor crítico de la distribución normal estándar
  z_critico = np.abs(norm.ppf((1 - confianza) / 2))
  # print (z_critico)
  z_critico_izq = -(z_critico)
  z_critico_der = z_critico
  #print (np.round(z_critico_izq, 6), np.round(z_critico_der, 6))

  # Valores de límites del intervalo de confianza
  # Calcular los límites del intervalo de confianza
  limite_inferior = media_m - z_critico * es
  limite_superior = media_m + z_critico * es

  intervalo = [limite_inferior, limite_superior]
  return intervalo

def f_visualizar_dens_area_normal_z(media_m, desv_std_p, n, confianza, titulo='Normal z'):

    # Crear un rango de valores para la PDF
    x = np.linspace(media_m - 3 *desv_std_p, media_m + 3 * desv_std_p, 1000)
    z = f_transformar_x_z(x, media_m, desv_std_p)

    # Determinar z criticos izquierda y derecha
    # es = desv_std_p / np.sqrt(n)  # error estándar desv_std/sqrt(n)

    # Calcular el valor crítico de la distribución normal estándar
    z_critico = np.abs(norm.ppf((1 - confianza) / 2))
    # print (z_critico)
    z_critico_izq = -(z_critico)
    z_critico_der = z_critico
    # Calcular la PDF de la distribución normal Z
    pdf = norm.pdf(z)


    # La probabilidad
    prob_izq  = norm.cdf(z_critico_izq) - norm.cdf(-np.inf)
    prob_der  = norm.cdf(np.inf) - norm.cdf(z_critico_der)
    # Visualizar la PDF
    plt.plot(z, pdf, label=titulo+':(me:{}; ds:{})'.format(0, 1))
    prob = prob_izq + prob_der
    # plt.fill_between(x, 0, pdf, where=(z >= np.inf) & (z <= z_critico_izq)
    plt.fill_between(z, 0, pdf, where=((z >= -np.inf) & (z <= z_critico_izq)) | ((z <= np.inf) & (z >= z_critico_der)), alpha=0.3, color='red', label='Prob='+str(np.round(prob,2))+"%")

    plt.axvline(z_critico_izq, color='red', linestyle='--', label='z crítico izq:'+str(np.round(z_critico_izq, 4))+';'+str(confianza*100)+'%')
    plt.axvline(z_critico_der, color='red', linestyle='--', label='z crítico der:'+str(np.round(z_critico_der, 4))+';'+str(confianza*100)+'%')
    plt.title('Función de Probabilidad (PDF) - '+titulo)
    plt.xlabel('Valor de z')
    plt.ylabel('Densidad ')
    plt.legend(loc='upper right')
    # plt.grid(True)
    #plt.show()

def f_visualizar_ic_normal(media_m, desv_std_p, n, confianza):
  # Generar datos para la distribución normal
  x = np.linspace(media_m - 1 * desv_std_p, media_m + 1 * desv_std_p, 1000)
  y = norm.pdf(x, loc=media_m, scale=desv_std_p / np.sqrt(n))

  intervalo = f_intervalo_confianza_z(media_m, desv_std_p, n, confianza)

  # Valores de límites del intervalo de confianza
  # Calcular los límites del intervalo de confianza
  limite_inferior = intervalo[0]
  limite_superior = intervalo[1]

  # Visualizar la distribución normal y el intervalo de confianza
  plt.plot(x, y, label='Distribución normal')
  plt.axvline(limite_inferior, color='red', linestyle='--', label='Límite Inf:'+str(np.round(limite_inferior, 2)))
  plt.axvline(limite_superior, color='red', linestyle='--', label='Límite Sup:'+str(np.round(limite_superior, 2)))
  plt.title(f'Confianza al {confianza * 100}% para Media Poblacional')
  plt.xlabel('Media Poblacional')
  plt.ylabel('Densidad de Probabilidad')
  plt.legend()
  # plt.show()



# Se reciben valores de x y se devuelven valores en z
def f_transformar_x_z(x, media, desv_std):
  x = (x - media) / desv_std
  return x

def f_intervalo_confianza_t(media_m, desv_std_m, n, confianza):

  gl = n-1
  # Calcular el error estándar de la media (SEM)
  es = desv_std_m / np.sqrt(n)  # error estándar desv_std/sqrt(n)
  #print(es)
  # Calcular el valor crítico de la distribución normal estándar
  t_critico = np.abs(t.ppf((1 - confianza) / 2, df = gl))
  # print (z_critico)
  t_critico_izq = -(z_critico)
  t_critico_der = z_critico
  #print (np.round(z_critico_izq, 6), np.round(z_critico_der, 6))

  # Valores de límites del intervalo de confianza
  # Calcular los límites del intervalo de confianza
  limite_inferior = media_m - t_critico * es
  limite_superior = media_m + t_critico * es

  intervalo = [limite_inferior, limite_superior]
  return intervalo

def f_visualizar_ic_t(media_m, desv_std_m, n, confianza):
  # Generar datos para la distribución normal
  x = np.linspace(media_m - 1 * desv_std_m, media_m + 1 * desv_std_m, 1000)
  y = norm.pdf(x, loc=media_m, scale=desv_std_m / np.sqrt(n))

  intervalo = f_intervalo_confianza_t(media_m, desv_std_m, n, confianza)

  # Valores de límites del intervalo de confianza
  # Calcular los límites del intervalo de confianza
  limite_inferior = intervalo[0]
  limite_superior = intervalo[1]

  # Visualizar la distribución normal y el intervalo de confianza
  plt.plot(x, y, label='Distribución t student')
  plt.axvline(limite_inferior, color='red', linestyle='--', label='Límite Inf:'+str(np.round(limite_inferior, 2)))
  plt.axvline(limite_superior, color='red', linestyle='--', label='Límite Sup:'+str(np.round(limite_superior, 2)))
  plt.title(f'Confianza al {confianza * 100}% para Media Poblacional')
  plt.xlabel('Media Poblacional')
  plt.ylabel('Densidad de Probabilidad')
  plt.legend()
  # plt.show()
```

## Valores críticos de z

Se presentan los valoes criticos de z a un $80$%, $90$%, $95$% y $95$% nivel de cinfianza.

Se utiliza la función *norm.ppf()* para devolver el valor de $z$ de acuerdo a esa probabilidad a ambas colas de la curva.




```python
confianza = [0.80, 0.90, 0.95, 0.99]
# Calcular el valor crítico de la distribución normal estándar
z_critico = np.abs(norm.ppf((1 - np.array(confianza)) / 2)) # dos colas
# print (z_critico)
z_critico_izq = -(z_critico)
z_critico_der = z_critico
confianzas_z = pd.DataFrame({'confianza':confianza, 'z critico':np.round(z_critico, 6),
                            'z crítico izq':np.round(z_critico_izq, 6), 'z crítico der':np.round(z_critico_der, 6)})
print ("Valor de nivels de confianza y valores de z a dos colas")
print (confianzas_z)
```

## Ejercicio

Se tiene una muestra de un contexto genérico de tamaño $n = 30$, una media muestral de $\bar{x}=100$, una desviación estándar poblacional conocida $ \sigma = 15$, y se desea calcular un intervalo de confianza del $95\%$.

$$
IC = \bar{x} \pm z \cdot \frac{\sigma}{\sqrt{n}} \therefore
$$

$$
IC = 100 \pm 1.96 \cdot \frac{15}{\sqrt{30}}
$$

$$
IC = 100 \pm 1.96 \cdot 2.7386
$$

$$
IC_{inf} = 100 - 1.96 \cdot 2.7386 = 94.63
$$

$$
IC_{sup} = 100 + 1.96 \cdot 2.7386 = 105.36
$$


### Inicializar valores

Se construyen los datos iniciales con las variables asignadas con los valores adecuados; luego se manda llamar la función *f_intervalo_confianza_z()* previamente codificada y se obtienen el intervalo de confianza.

Dado que se conoce la desviación estándar de la población se utiliza la distribución normal $z$.

```python
# Inicializar valores
n = 30
media_m = 100
desv_std_p = 15
confianza = 0.95 # 95%
alfa = (1 - confianza) / 2 # a dos colas

z_critico = np.abs(norm.ppf((1 - confianza) / 2))
# print (z_critico)
z_critico_izq = -(z_critico)
z_critico_der = z_critico

datos = pd.DataFrame({'n':[n], 'media_m':[media_m], 'desv std p':[desv_std_p],
                      'confianza':[confianza,], 'alfa':[alfa],
                      'z crítico izq':np.round(z_critico_izq, 6),
                      'z crítico der':np.round(z_critico_der, 6)})

print ("Datos iniciales")
print (datos)
```

### Calcular intervalo de confianza

```python
print ("Intervalo de confianza de la media poblacional ")
intervalo = f_intervalo_confianza_z(media_m, desv_std_p, n, confianza)
print (np.round(intervalo, 4))
```

## Visualizando el intervalo de confianza



```python
plt.figure(figsize=(14, 4))
plt.subplot(1, 2, 1)  # Primer subgráfico
f_visualizar_dens_area_normal_z(media_m, desv_std_p, n, confianza, titulo='normal z')
plt.subplot(1, 2, 2)  # Segundo subgráfico
f_visualizar_ic_normal(media_m, desv_std_p, n, confianza)
plt.show()
```

El intervalo de confianza estimado de la media poblacional debe estar entre 94.63 y 105.36.

## Valores críticos de t

Se presentan los valores críticos de $t$ a un $80$%, $90$%, $95$% y $95$% nivel de cinfianza.

Se utiliza la función *t.ppf()* para devolver el valor de $t$ de acuerdo a esa probabilidad a ambas colas de la curva.

La función t.ppf() requiere los grados de libertad necesarios para tratar con *t student* por lo que se le incorpora por esta ocasión $29$ dado que una muestra pequeña se considera de $n=30$ observaciones y grados de libertad es $gl = n-1$


```python
confianza = [0.80, 0.90, 0.95, 0.99]
n = 30
gl = n - 1
# Calcular el valor crítico de la distribución t student a 29 gl
t_critico = np.abs(t.ppf((1 - np.array(confianza)) / 2, df= gl)) # dos colas

# print (z_critico)
t_critico_izq = -(t_critico)
t_critico_der = t_critico
confianzas_t = pd.DataFrame({'confianza':confianza, 't critico':np.round(t_critico, 6),
                            't crítico izq':np.round(t_critico_izq, 6), 't crítico der':np.round(t_critico_der, 6)})
print ("Valor de niveles de confianza y valores de t; a ", gl, "grados de libertad a dos colas")
print (confianzas_t)
```

## Ejercicio del ambientalista

Un ambientalista está realizando un estudio del oso polar, especie que se encuentra en el océano Ártico y sus alrededores. Su zona de distribución está limitada por la existencia de hielo en el mar, que usan como plataforma para cazar focas, principal sostén de los osos.

La destrucción de su hábitat en el hielo del Ártico, que se ha atribuido al calentamiento global, amenaza la supervivencia de los osos como especie; puede extinguirse antes de un siglo.

Una muestra aleatoria de $n=50$ osos polares produjo un peso promedio de $\bar{x}=980$ libras con una desviación estándar de $s=105$ libras.

### Datos iniciales


```python
# Inicializar valores
n = 50
gl = n-1
media_m = 980
desv_std_m = 105
confianza = 0.90 # 90%
alfa = (1 - confianza) / 2 # a dos colas

t_critico = np.abs(t.ppf((1 - confianza) / 2, df = gl))
# print (z_critico)
t_critico_izq = -(t_critico)
t_critico_der = t_critico

datos = pd.DataFrame({'n':[n], 'media ':[media_m], 'desv std':[desv_std_m],
                      'confianza':[confianza,], 'alfa':[alfa],
                      't crítico izq':np.round(t_critico_izq, 6),
                      't crítico der':np.round(t_critico_der, 6)})
print ("Datos iniciales")
print (datos)
```

# Interpretación

La relación entre el nivel de confianza y el ancho del intervalo de confianza es directa. A mayor nivel de confianza, se requiere un intervalo de confianza más amplio.

Cuando se construye un intervalo de confianza para la media poblacional utilizando la distribución normal z, se está expresando la certeza (o confianza) de que el parámetro de interés (la media) está contenido dentro de ese intervalo.

Un intervalo de confianza más amplio indica una mayor certeza, pero a costa de proporcionar menos precisión sobre la ubicación exacta de la media.

Por ejemplo, si se elige un nivel de confianza del $95$%, el intervalo de confianza contendrá la verdadera media con un $95$% de probabilidad, pero el intervalo puede ser más amplio.

Si se elige un nivel de confianza del $99$%, el intervalo será más amplio para garantizar un mayor grado de certeza. En resumen, a mayor nivel de confianza, mayor será el intervalo de confianza.
