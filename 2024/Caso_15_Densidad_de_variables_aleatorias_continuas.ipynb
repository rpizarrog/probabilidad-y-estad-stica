# Caso 15 Densidad y probabilidad de variables aleatorias continuas

En el contexto de la teoría de probabilidad, las variables aleatorias continuas son aquellas que pueden tomar valores en un rango continuo de números reales.

A diferencia de las variables aleatorias discretas, que solo pueden tomar valores en un conjunto discreto de puntos, las variables aleatorias continuas pueden tomar una cantidad infinita de valores en un intervalo.

Los conceptos de variable aleatoria continua y de función de densidad se definen a partir de la noción de probabilidad. Como una mera aproximación, se dice que la función de densidad sería el polígono de frecuencias que se construiría si se tuviese un conjunto infinito de datos.

## La Función de Densidad de Probabilidad (PDF):

La PDF es una función que describe cómo se distribuyen las probabilidades en una variable aleatoria continua a lo largo de su rango.

La integral $\int$ de la PDF sobre un intervalo específico proporciona la probabilidad de que la variable aleatoria caiga dentro de ese intervalo.

La funcion de densidad de probabilidad se denota por medio de $f(x)$.

Algunas propiedades de la densidad de probabilidad (PDF) son las siguientes:

  - Es no negativa:$f(x) \ge 0$
  - La integral sobre todo el rango es igual a 1:

    $$
    \int_{-∞}^{∞} f(x)dx = 1
    $$
  
    Toda el área bajo la curva es el 100% en términos porcentuales o 1 en términos relativos.

## Función de Distribución Acumulativa (CDF):

La CDF es una función que proporciona la probabilidad acumulativa de que una variable aleatoria sea menor o igual a un valor específico. Se denota como:
$$
F(x) = \int_{-∞}^{x}f(x)dx
$$

donde $f(x)$ es la PDF.

## Valor Esperado (Media)

El valor esperado de una variable aleatoria continua se calcula como el promedio ponderado de todos los valores posibles, donde el peso es dado por la PDF. Matemáticamente, se expresa como
$$
\mu_{x} = VE(x) = \int_{-\infty}^{\infty}x\cdot f(x)dx
$$

## Varianza y Desviación Estándar:

La varianza y la desviación estándar miden la dispersión de una variable aleatoria continua. La varianza se calcula como:

$$
varianza = \sigma^2=V(x)=\int_{-\infty}^{\infty}(x - \mu)^2\cdot f(x) dx
$$
, donde $\mu$ es la media, y la desviación estándar es la raíz cuadrada de la varianza

$$
\sigma = \sqrt{\sigma^2}
$$

Las variables aleatorias continuas son esenciales para modelar una amplia gama de fenómenos en ciencia, ingeniería, estadísticas y muchas otras disciplina, entre otros.

```python
import numpy as np
import matplotlib.pyplot as plt
import math

from scipy.stats import uniform
from scipy.stats import norm
from scipy.stats import chi2
```

## Crear funciones

### Cuadradro de x menos 2
$$
f(x) = x^{2} - 2
$$

### Cubo de x + 1
$$
f(x) = x^3 + 1
$$

### Uniforme
$$
f(x) = \frac{1}{b-a}
$$

### Normal
$$
f(x) =\frac{1}{\sigma \cdot \sqrt{2 \cdot \pi}}\cdot e
^{\frac{-1}{2}\cdot ( \frac {x - \mu}{\sigma}) ^2}
$$

### Chi cuadrada
$$
f(x) = \frac{1}{2^{k/2} * \gamma^{(k/2)}} *
e^{(-x/2)} * x^{(k/2) - 1)}  \therefore
$$

o bien:
$$
f(x) = \frac{x^{(k/2 - 1)} * e^{(-x/2)}}{2^{(k/2)} * \gamma(k/2)}
$$

$$
\gamma \text{:la función gamma} \\
e \text{ el valor de e = 2.718282}
$$

$$
\quad x \in (0, \infty)
$$



```python
# Definir la función de densidad
def f_x_cuad_menos_2(x):
  str_fx = "f(x) = x ^ 2 -2"
  fx = x**2 - 2
  return fx, str_fx

def f_x_cubo_mas_1(x):
  str_fx = "f(x) = x ^ 3 + 1"
  fx = x ** 3 + 1
  return fx, str_fx

def f_visualizar_densidad (funcion, x):
  # Crear un rango de valores para la PDF

  # Calcular la PDF (probability density function)
  dens_pdf, str_fx = funcion(x)

  # Visualizar la PDF
  plt.plot(x, dens_pdf, label='PDF: f(x) = ' + str_fx)
  plt.title('Función de Densidad de Probabilidad (PDF)')
  plt.xlabel('Valor de x')
  plt.ylabel('Densidad de Probabilidad')
  plt.legend()
  # plt.grid(True)
  plt.show()

# Densidad uniforme
def f_x_uniforme (x, a, b):
  if a <= x <= b:
    return 1 / (b - a)
  else:
    return 0

def f_visualizar_dens_uniforme (x, a, b):
  str_fx = "f(x) = 1 / (b - a)"
  # La densidad

  # Calcular la PDF para cada valor en el rango
  dens_pdf = [f_x_uniforme(xi, a, b) for xi in x]

# Visualizar la PDF
  plt.plot(x, dens_pdf, label='PDF: f(x) = '+ str_fx)
  plt.title('Función de Densidad de Probabilidad (PDF)')
  plt.xlabel('Valor de x')
  plt.ylabel('Densidad de Probabilidad')
  plt.legend()
  # plt.grid(True)
  plt.show()

# Densidad normal
def f_x_normal(x, media, desviacion_estandar):
  # Calcular la PDF de la variable aleatoria continua
  fx = 1 / (desviacion_estandar * np.sqrt(2 * np.pi)) * np.exp(-(x - media)**2 / (2 * desviacion_estandar**2))
  return fx,

# Función para visualizar la densidad de probabilidad de una distribución normal
def f_visualizar_dens_normal(valores, media, desviacion_estandar):
    # Crear un rango de valores para la PDF
    x = np.linspace(media - 3 * desviacion_estandar, media + 3 * desviacion_estandar, 1000)

    # Calcular la PDF de la distribución normal
    pdf = 1 / (desviacion_estandar * np.sqrt(2 * np.pi)) * np.exp(-(x - media)**2 / (2 * desviacion_estandar**2))

    # Visualizar la PDF
    plt.plot(x, pdf, label='PDF: Normal({},{})'.format(media, desviacion_estandar))
    plt.title('Función de Densidad de Probabilidad (PDF) - Distribución Normal')
    plt.xlabel('Valor de x')
    plt.ylabel('Densidad de Probabilidad')
    plt.legend()
    # plt.grid(True)
    plt.show()

# Densidad chi cuadrada
def f_x_dens_chi_cuadrada(x, k):
  # Calcula la densidad de una chi cuadrada
  # recibe los valores de x y los grados de libertad (n-1)
  if x < 0:
    return 0
  numerador = x**((k / 2) - 1) * np.exp(-x / 2)
  denominador = (2**(k / 2)) * math.gamma(k / 2)
  dens_pdf = numerador / denominador
  return dens_pdf


def f_visualizar_dens_chi_cuadradada (gl):
  # Crear un rango de valores para la PDF
  x = np.linspace(0, 20, 1000)

  # Calcular la PDF para cada valor en el rango
  dens_pdf = [f_x_dens_chi_cuadrada(xi, gl) for xi in x]

  # Visualizar la PDF
  plt.plot(x, dens_pdf, label=f'PDF: Chi-cuadrado({gl})')
  plt.title('Función de Densidad de Probabilidad (PDF) - Distribución Chi-cuadrado')
  plt.xlabel('Valor de x')
  plt.ylabel('Densidad de Probabilidad')
  plt.legend()
  plt.show()

```

```python
x = 30
a = 20
b = a + 30
n = 100

# Calcula la densidad con la función personalizada f_x_uniforme()
dens_pdf = f_x_uniforme(x, a, b)
print("Densidad para un valor de x igual a ", x, dens_pdf)

# Calcula la densidad con uniform.pdf de librería scypi.stats
pdf = uniform.pdf(x, loc=a, scale=b-a)
print("Densidad para un valor de x igual a ", x, pdf)

# Visualizar la densidad de valores de 20 a 40
x = np.linspace(0, 60, n) # Números entre 0 y 50
f_visualizar_dens_uniforme(x, a=a, b=b) # intervalo entre 20 y 40

```
