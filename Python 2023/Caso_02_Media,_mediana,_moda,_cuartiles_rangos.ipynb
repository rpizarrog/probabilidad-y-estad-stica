## Cargar librerías



```python
import numpy as np
import pandas as pd
import statistics
import matplotlib.pyplot as plt

import seaborn as sns # Para graficar histograma con densidad
```

## Crear los datos




```python
numeros = [40, 60, 50, 45, 65, 70, 95, 90, 45, 60, 43, 56, 65, 80, 45, 70, 45, 75, 45, 54, 35, 46, 47, 50, 50, 60, 50, 50, 65, 50, 50, 22, 54, 68, 70, 46, 54, 55, 50, 55, 40, 68, 76, 56, 55, 45, 50, 43, 46, 47, 70, 24, 34, 54, 43, 34, 45, 45, 45, 45]
print(numeros)
print ("Números ordenados")
print(sorted(numeros))
```

## Crear estadísticos

Se calcula la media artimética, la mediana, la moda, los cuartiles, el rango de los datos


```python
media = np.mean(numeros)
mediana = np.median(numeros)
moda = statistics.mode(numeros)
maximo = max(numeros)
minimo = min(numeros)
rango = maximo - minimo
print("La media es:", round(media, 4))
print("La mediana es:", round(mediana, 4))
print("La moda es:", round(moda, 4))
print("El valor mínimo es:", round(minimo, 4))
print("El valor máximo es:", round(maximo, 4))
print("El rango es: ", rango)

```

¿Son la media y la mediana iguales?


```python
if media == mediana:
  print ("La media y mediana son iguales, de tal forma que los datos presentan una condicón simétrica")
else:
  print ("La media y mediana son diferentes, de tal forma que los datos presentan una condicón asimétrica")
```

### La descripción de los datos

Convertir los números a un data.frame para utilziar describe()

```python
datos = pd.DataFrame(numeros, columns=["numeros"])
datos
```

```python
datos.describe()
```

### Hisograma sin la densidad

```python
plt.hist(numeros, bins=10, alpha=0.5, color='b', edgecolor='black')
plt.axvline(media, color='r', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(mediana, color='g', linestyle='dashed', linewidth=2, label='Mediana')
plt.axvline(moda, color='y', linestyle='dashed', linewidth=2, label='Moda')
plt.legend()
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma de Datos con Media, Mediana y Moda')
plt.show()
```

### Histograma de los numeros con densidad


```python
sns.histplot(x = numeros, kde = True,
             kde_kws = {'bw_adjust': 0.5})
```

Se observa que los datos presentan una distribución asimétrica con sesgo a la izquierda.

## Cuartiles de los números

Los cuartiles representan los valores que están a las cuartas partes de los datos:
* Primer cuartil, el valor que está a la primera caurta parte de los datos ordenados.
* Segundo cuartil es la mediana o el valor que está a la segunda caurta parte de los datos ordenados.
* Tercer cuartil es el valor que está a la tercera cuarta parte de los datos ordenados.

Luego:

El rango intercuartílico (RI) es la diferencia entre el tercer y primer cuartil

* Los valores que están por encima de tercer cuartil mas 1.5 veces el RI son datos atípicos y el diagrama de caja ayuda a encontrarlos.

* Los valores que están por debajo del primer cuartil menos 1.5 veces el RI son datos atípicos y el diagrama de caja ayuda a encontrarlos.




```python
# caurtiles
cuartiles = datos.quantile([0.25, 0.5, 0.75])
print(cuartiles)
#

q1 = np.percentile(datos, 25)  # Primer cuartil (Q1)
q2 = np.percentile(datos, 50)  # Segundo cuartil (Q2, mediana)
q3 = np.percentile(datos, 75)  # Tercer cuartil (Q3)

print("Primer cuartil (Q1):", q1)
print("Segundo cuartil (Q2, Mediana):", q2)
print("Tercer cuartil (Q3):", q3)

# calcular el rango intercuartílico
RI = q3 - q1
print("Rango intercuarílico (RI):" , RI)


print ("Los valores que están por debajo de ", q1 - 1.5 * RI, " son atípicos")
print ("Los valores que están por encima de ", q3 + 1.5 * RI, " son atípicos")
```

### Diagrama de caja


```python
plt.boxplot(numeros)
plt.xlabel('Conjunto de Datos')
plt.ylabel('Valor')
plt.title('Diagrama de Caja de Datos')
plt.show()
```

Existen tres datos atípicos

* Hay dos datos por encima de el tercer cuartil mas 1.5 veces el rango intercuartílico. 22.5

* Hay un dato por debajo del prier cuartil menos 1.5 veces el rango intercuartilico. 82.5.
