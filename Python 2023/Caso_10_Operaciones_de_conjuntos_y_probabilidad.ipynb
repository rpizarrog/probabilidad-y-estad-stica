# Marco teórico

## Operación Union U.

El conjunto de todos los elementos que pertenecen a A o a B, o tanto a A como a B, se llama la unión de A y B y se escribe A ∪ B.

En la unión si hay elementos repetidos sólo se deja uno de ellos.

![](images/union%20conjuntos.png)

## Operación intersección ∩

El conjunto de todos los elementos que pertenecen simultáneamente a A y B se llama la intersección de A y B y se escribe A ∩ B.

![](images/interseccion-01.png)

## Operación diferencia -

El conjunto que consiste en todos los elementos de A que no pertenecen a B se llama la diferencia de A y B y se escribe A -- B.

![](images/diferencia.png)

## Operación complemento [C ó ´]

Son todos los conjuntos con los elementos que no están en A y se escribe A' ó C A. Son todos los elementos que faltan y que no están en A para complementar todo el espacio muestral.

![](images/complemento-01.png)

# Desarrollo

## Cargar librerías

```python
# Cargar librerías
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
from matplotlib_venn import venn3
```

## Cargar datos


```python
url = "https://raw.githubusercontent.com/rpizarrog/probabilidad-y-estad-stica/master/2023/datos/alumnos_actividades_extraescolares.csv"
datos = pd.read_csv(url)



print("Los nombres de datos (alumnos)")

print(datos.head(70))
```

## Crear conjuntos de varias disciplinas
* FU : FUTBOL
* BA : BASQUETBOL
* VO : VOLEYBOL
* TE : TENIS
* AJ : AJEDREZ
* AT : ATLETISMO
* DA : DANZA
* RO : RONDALLA
* TEA : TEATRO





```python
random.seed(2023)
nombres = datos["nombres"].tolist()
N = len(nombres)

# Barajar los nombres aleatoriamente
random.shuffle(nombres)

FU = set(random.sample(nombres,  20))
print("Conjunto FU: FUTBOL")
print(FU)
n = len(FU)
print("La probabilidad de elegir a un alumno que sea de FUTBOL es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")
print("\n")

BA = set(random.sample(nombres,  12))
print("Conjunto BA: BASQUETBOL")
print(BA)
n = len(BA)
print("La probabilidad de elegir a un alumno que sea de BASQUETBOL es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")
print("\n")


VO = set(random.sample(nombres,  12))
print("Conjunto VO: VOLEYBOL")
print(VO)
n = len(VO)
print("La probabilidad de elegir a un alumno que sea de VOLEYBOL es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")
print("\n")

TE = set(random.sample(nombres,  10))
print("Conjunto TE: TENIS")
print(TE)
n = len(TE)
print("La probabilidad de elegir a un alumno que sea de TENIS es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")
print("\n")

AJ = set(random.sample(nombres,  8))
AJ
print("Conjunto AJ: AJEDREZ")
print(AJ)
n = len(AJ)
print("La probabilidad de elegir a un alumno que sea de AJEDREZ es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")
print("\n")

AT = set(random.sample(nombres,  16))
print("Conjunto AT: ATLETISMO")
print(AT)
n = len(AT)
print("La probabilidad de elegir a un alumno que sea de ATLETISMO es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")
print("\n")

DA = set(random.sample(nombres,  12))
DA
print("Conjunto DA: DANZA")
print(DA)
n = len(DA)
print("La probabilidad de elegir a un alumno que sea de DANZA es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")
print("\n")

RO = set(random.sample(nombres,  14))
print("Conjunto RO: RONDALLA")
print(RO)
n = len(RO)
print("La probabilidad de elegir a un alumno que sea de RONDALLA es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")
print("\n")

TEA = set(random.sample(nombres,  9))
print("Conjunto TEA: TEATRO")
print(TEA)
n = len(DA)
print("La probabilidad de elegir a un alumno que sea de TEATRO es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")
print("\n")

```

## Unión de conjuntos U

### Unión de FUTBOL con ATLETISMO




```python
# FUTBOL U ATLETISMO
resultado = FU.union(AT)
print(resultado)
n = len(resultado) # resulado de la operación de conjuntos
print("La probabilidad de elegir a un alumno que sea de FUTBOL O ATLETISMO es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")


# Diagrama de Venn
venn2([FU, AT], ('FUTBOL', 'ATLETISMO'))

# Mostrar los elementos de cada conjunto en el título
plt.title("Unión de Conjuntos FUTBOL y ATLETISMO\nFU: {}\nAT: {}".format(FU, AT))
plt.show()
```

### Unión de TEATRO con RONDALLA


```python
# TEATRO U RONDALLA
resultado = TEA.union(RO)
print(resultado)
n = len(resultado) # resulado de la operación de conjuntos
print("La probabilidad de elegir a un alumno que sea de TATRO O RONDALLA es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")


# Diagrama de Venn
venn2([TEA, RO], ('TEATRO', 'RONDALLA'))

# Mostrar los elementos de cada conjunto en el título
plt.title("Unión de Conjuntos TEATRO y RONDALLA\nTEA: {}\nRO: {}".format(TEA, RO))
plt.show()
```

## AJEDREZ U TENIS


```python
# AJEDREZ U TENIS
resultado = AJ.union(TE)
print(resultado)
n = len(resultado) # resulado de la operación de conjuntos
print("La probabilidad de elegir a un alumno que sea de AJEDREZ O TENIS es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")

# Diagrama de Venn
venn2([AJ, TE], ('AJEDREZ', 'TENIS'))

# Mostrar los elementos de cada conjunto en el título
plt.title("Unión de Conjuntos AJEDREZ y TENIS\nAJ: {}\nTE: {}".format(AJ, TE))
plt.show()
```

### Intersección de conjuntos ⋂
#### Intersección  de FUTBOL con ATLETISMO

```python
# FUTBOL ⋂ ATLETISMO
resultado = FU.intersection(AT)
print(resultado)
n = len(resultado) # resulado de la operación de conjuntos
print("La probabilidad de elegir a un alumno que sea de FUTBOL y ATLETISMO es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")

# Diagrama de Venn
venn2([FU, AT], ('FUTBOL', 'ATLETISMO'))

# Mostrar los elementos de cada conjunto en el título
plt.title("Unión de Conjuntos FUTBOL y ATLETISMO\nFU: {}\nAT: {}".format(FU, AT))
plt.show()
```

### Intersección de BASQUETBOL con VOLEYBOL

```python
# BASQUETBOL ⋂ VOLEYBOL
resultado = BA.intersection(VO)
print(resultado)
n = len(resultado) # resulado de la operación de conjuntos
print("La probabilidad de elegir a un alumno que sea de BASQUETBOL y VOLEYBOL es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")

# Diagrama de Venn
venn2([BA, VO], ('BASQUETBOL', 'VOLEYBOL'))

# Mostrar los elementos de cada conjunto en el título
plt.title("Unión de Conjuntos BASEUTBOL y VOLEYBOL\nBA: {}\nVO: {}".format(BA, VO))
plt.show()
```

# Intersección AJEDREZ, RONDALLA Y ATLETISMO

¿cuál es la probabildia de elegir a un alumno que esté inscritp en AJEDREZ RONDALLA Y TEARO?


```python
# AJEDREZ ⋂ RONDALLA ⋂ TEATRO
resultado = (AJ.intersection(RO)).intersection(TEA)
print(resultado)
n = len(resultado) # resulado de la operación de conjuntos
print("La probabilidad de elegir a un alumno que sea de AJEDREZ, RONDALLA Y TEATRO es: ",n,"/",N,"=",np.round(n/N, 4), "≈", np.round(n/N * 100, 2), "%")

print(AJ)
print(RO)
print(TEA)
# Diagrama de Venn
venn3([AJ, RO, TEA], ('AJEDREZ', 'RONDALLA', 'TEATRO'))

# Mostrar los elementos de cada conjunto en el título
plt.title("Unión de Conjuntos AJEDREZ, Y RONDALLA Y TEATRO\nAJ: {}\nRO: {}\nTEA: {}".format(AJ, RO, TEA))
plt.show()
```
