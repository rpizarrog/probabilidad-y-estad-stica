# Objetivo

Realizar permutaciones sin repetición y encontrar probabilidades con la suma de dos cartas de entre todo el espacio muestral.

# Descripción

Se tiene una baraja de 16 cartas con valores de entre 1, 2, 3, 4 cartas de cada figura de entre A, B, C, D.

Las cartas:
{
  A1, B1, C1, D1,
  A2, B2, C2, D2,
  A3, B3, C3, D3,
  A4, B4, C4, D4
}

Son 16 cartas de este universo de baraja.

Se reparten dos cartas y se contabilizan los puntos.

Por ejemplo:

* La carta A1 vale 1 punto,
* La carta A2 vale 2 punto,
* La carta B2 vale 2 puntos,
* La carta C3 vale 3 puntos,
* La carta D4 vale 4 puntos,
* Si se reparten las cartas A4 y B2 serían 6 puntos,
* Si se reparten las cartas A2 y B2 serían 4 puntos,
* Si se reparten las cartas C4 y C1 serían 5 puntos.

**Se pide:**
Se debe crear las posibles permutaciones sin repetición y construir el espacio muestral de las 4 cartas y 4 figuras de cada una.

$$
Pr = \frac{n!}{(n-1)!}
$$

A partir de las posibles permutaciones $Pr=240$, se debe construir la tabla de frecuencias con las sumas desde 0 hasta 8 que son las sumas posibes de tener dos cartas.

De la tabla de frecuencias responder a las preguntas de probabilidad:

* Determinar y calcular el número de permutaciones sin repetición: Deben ser 240 : n! / (n-r)!
* ¿Cuál es la número de posibles permutaciones sin repetición? n = 240?

* Elaborar un diagrama de barra que identifique las probabilidades de cada valor de las sumas.  
Por supuesto no hay probabilidad de que la suma sea 0, 1, 9 o 10, porque no hay dos cartas que la suma de esos valores sume dicha cantidad.
* ¿Cúal es la probabilidad de si se reparten dos cartas la suma de estas cartas sea 4?, ¿porqué?
* ¿Cuál es la probabilidad de que si se reparten dos cartas la suma de estas cartas sea 8?, ¿porqué?
* ¿Qué es más probable que la suma sea un 4 o que la suma sea un 5?, ¿porqué?
* ¿Qué es má probables que la suma sea 3 o 5 o que la suma sea 6 o 7?, ¿porqué?

# Las librerías

```python
# Librerías
# Cargar librerías
import numpy as np
import math as mt # Para utilizar factorial
from itertools import permutations
from itertools import product
from itertools import combinations
from itertools import combinations_with_replacement
import pandas as pd
import matplotlib.pyplot as plt
```

# Construir espacio y tabla de frecuencias

* Se declara el universo de cartas
* Se contruye el espacio muestral
* Se construye la tabla de distribución

```python
# Permutaciones
# Formar las permutaciones sin repetición
baraja = ['A1', 'A2', 'A3', 'A4',
          'B1', 'B2', 'B3', 'B4',
          'C1', 'C2', 'C3', 'C4',
          'D1', 'D2', 'D3', 'D4']
n = len(baraja)
r = 2 # ¿cómo agrupar?, de dos en dos porque se reparten dos barajas

n_PrR = mt.factorial(n) / (mt.factorial(n - r))

print("Existen ", n_PrR , " posibles permutaciones con repetición del total de ", n , " al seleccionar dos barajas de tipo inglesa")

print("Aquí las permutaciones con repetición de dos barjas y la suma de los puntos")

# Calcular las permutaciones con repetición
# permutaciones = product(baraja, repeat=r)
```

## Las permutaciones


```python
permutaciones = list(permutations(baraja, 2))
permutaciones
```

## Las sumas


```python
from numpy.lib.function_base import append
# Función para calcular la suma de los valores numéricos de dos cartas
def calcular_suma(carta1, carta2):
    numeros_carta1 = int(carta1[1:])
    numeros_carta2 = int(carta2[1:])
    return numeros_carta1 + numeros_carta2

# Convertir a data.frame con pandas pd
tabla = pd.DataFrame(permutaciones, columns=['Carta1', 'Carta2'])
# print(tabla)

# Agregar la columna suma
suma = []
for i in range(len(tabla)):
  suma.append(calcular_suma(tabla.loc[i, 'Carta1'], tabla.loc[i, 'Carta2']))

tabla['suma'] = suma

# Primeras 50
print ("Primeras 50")
tabla_subset = tabla.iloc[0:50]
print(tabla_subset)

# Siguientes 50-100
print ("Siguientes 50")
tabla_subset = tabla.iloc[50:100]
print(tabla_subset)

# Siguientes 50
print ("Siguientes 50")
tabla_subset = tabla.iloc[100:150]
print(tabla_subset)

# Siguientes 50
print ("Siguientes 50")
tabla_subset = tabla.iloc[150:200]
print(tabla_subset)

# Siguientes 40
print ("Siguientes 40")
tabla_subset = tabla.iloc[200:240]
print(tabla_subset)

#print(tabla_subset)

```

## Construyendo tabla de frecuencias


```python
# Calcular la frecuencia de cada valor en la columna "suma"
n = len(tabla)

tabla_frecuencias = tabla['suma'].value_counts().reset_index()
tabla_frecuencias.columns = ['suma', 'frecuencia']
# print(tabla_frecuencias)

# Ordenar la tabla de frecuencias por el valor
tabla_frecuencias = tabla_frecuencias.sort_values(by='suma')
# print(tabla_frecuencias)


# Definir la lista de nombres personalizados
nombres_indices = [0, 1, 2, 3, 4, 5, 6]

# Asignar los nuevos nombres de índice a la tabla de frecuencias
tabla_frecuencias.index = nombres_indices

tabla_frecuencias['Frec_rel'] = tabla_frecuencias['frecuencia'] / n
tabla_frecuencias['Frec_porc'] = tabla_frecuencias['frecuencia'] / n * 100
# Imprimir la tabla de frecuencias con los nuevos nombres de índice
print(tabla_frecuencias)
#print(tabla_frecuencias.to_markdown())
```

## Diagrama de barra

Se visualzia un diagrama de barra que representa de frecuencia de cada suma al repartirse dos cartas.



```python
import matplotlib.pyplot as plt

# Código para crear la tabla de frecuencias

# Crear un gráfico de barras
plt.figure(figsize=(8, 6))
bars = plt.bar(tabla_frecuencias['suma'], tabla_frecuencias['Frec_porc'])
plt.xlabel('Suma')
plt.ylabel('Frecuencia %')
plt.title('Diagrama de Barras de Probabilidad % por Suma de puntos de dos cartas')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Agregar los valores numéricos en las barras
# Agregar etiquetas de datos en cada barra
for x, y in zip(tabla_frecuencias['suma'], tabla_frecuencias['Frec_porc']):
    plt.text(x, y, f'{y:.2f}', ha='center', va='bottom')


# Mostrar el gráfico
plt.show()

```
